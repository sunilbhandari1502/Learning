```python
import pandas as pd
from nsepy import get_history
from datetime import date
import numpy as np 
import matplotlib.pyplot as plt  
%matplotlib inline
from statsmodels.api import OLS
from statsmodels.tsa.stattools import adfuller
from concurrent import futures
import talib as ta
import datetime
import warnings ; warnings.simplefilter("ignore")
from pandas_datareader import data
from math import *
from datetime import timedelta
from nsepy.derivatives import get_expiry_date
```

```python
def mergefunc(df,m): #Merging two dataframes
    mergef = pd.merge(df,m,how="inner",left_index = True, right_index = True)
    return mergef
```

```python

def tot_profit(m,SL,TP):
    positions = []
    exit = 0
    tradenotp = 0
    tradenosl = 0
    tradenomat = 0
    tradenomatp = 0
    count = 0
    
    for j in range(0,len(m)):
        #print("Going here")
        if m['Avgs'].iloc[j] > TP and exit == 0:
            tradenotp = tradenotp+1
            exit = 1
            #print("Trade no {x} .TP hit on date {y} value is {z} .".format
                      #(x=tradenotp, y=m.index[j], z = m['Avgs'].iloc[j]))
            positions.append(m['Avgs'].iloc[j])
        elif m['Avgs'].iloc[j] < SL and exit == 0:
            tradenosl = tradenosl+1
            exit = 1
            #print("Trade no {x} .SL hit on date {y} value is {z} .".format
                      #(x=tradenosl, y=m.index[j], z = m['Avgs'].iloc[j]))
            positions.append(m['Avgs'].iloc[j])
        elif j+1 < len(m):
            if (m['Expiry'].iloc[j] == m['Expiry'].iloc[j+1]):
                pass
            else:
                if exit == 0 :
                    tradenomat = tradenomat+1
                    #print("Trade no {x} .Maturity reched without anything hit on date {y} value is {z} .".format
                      #(x=tradenomat, y=m.index[j], z = m['Avgs'].iloc[j]))
                    positions.append(m['Avgs'].iloc[j])
                    if m['Avgs'].iloc[j] > 0 :
                        tradenomatp = tradenomatp + 1
                else:    
                     exit = 0
        else:
            continue
    tot_profit.Ntp = tradenotp
    tot_profit.Nsl = tradenosl
    tot_profit.Nmat = tradenomat
    tot_profit.Nmatp = tradenomatp
    return sum(positions)
```

```python

nifty_master = get_history(symbol="NIFTY",
                            start=date(2006,12,28),
                            end=date(2018,10,31),
                            index=True
                      )

nifty_master
```

```python
# Initial code for first month only
nifty_master["PDate"] = nifty_master.index

# Enter values here
yr = 2010
#######


x1 = get_expiry_date(year=yr-1, month=12)
x = nifty_master['PDate'].shift(-1)[nifty_master.index == x1].astype(date).iloc[0]

for i in range(1,2):
    exp = get_expiry_date(year=yr, month=i)
    nifty_ul = get_history(symbol="NIFTY",
                        start=x,
                        end=date(yr,i,7),
                        futures=True,
                        index=True,
                        expiry_date = exp)
    
    strk = (round(nifty_ul[nifty_ul.columns[2]]/100 ,0)*100).iloc[0]
    
    df1 = get_history(symbol="NIFTY",
                        start=x,index = True,
                        end=exp,
                        option_type="PE",
                        strike_price = strk,
                        expiry_date=exp)[["Expiry","Option Type","Strike Price","Open","High","Low","Close"]]
    df2 = get_history(symbol="NIFTY",
                        start=x,index = True,
                        end=exp,
                        option_type="CE",
                        strike_price = strk,
                        expiry_date=exp)[["Option Type","Open","High","Low","Close","Underlying"]]
    
    x = nifty_master['PDate'].shift(-1)[nifty_master.index == exp].astype(date).iloc[0]
    df = mergefunc(df1,df2)
    df["Cost"] = df["Open_x"].iloc[0] + df["Open_y"].iloc[0]
    df["Value"] = df["Low_x"] + df["Low_y"]
    df["Max_Profit"] = (df["Value"].max(axis=0)) - df['Cost'] 
    df["Max_DD"] = (df["Value"].min(axis=0)) - df['Cost'] 
    df["Max_Profit_Pct"] = df["Max_Profit"]/df["Cost"]
    df["Max_DD_Pct"] = df["Max_DD"]/df["Cost"]
df
```

```python

#Second code for remaining maturities

for j in range(yr,yr+1):
    for i in range(2,13):
        #exp = get_expiry_date(year=j, month=i)
        try:
            x = nifty_master['PDate'].shift(-1)[nifty_master.index == exp].astype(date).iloc[0]
        except:   
            exp = get_expiry_date(year=j, month=i)
            print("Error downloading month ", i)
            continue
            
        if (j == 2007 and i == 1) or (j==2018 and i>10) :
            continue
        
        else:
            
            exp = get_expiry_date(year=j, month=i)
            try:
                nifty_ul = get_history(symbol="NIFTY",
                        start=x,
                        end=date(j,i,7),
                        futures=True,
                        index=True,
                        expiry_date = exp)
    
                strk = (round(nifty_ul[nifty_ul.columns[2]]/100 ,0)*100).iloc[0]
    
                df1 = get_history(symbol="NIFTY",
                        start=x,index = True,
                        end=exp,
                        option_type="PE",
                        strike_price = strk,
                        expiry_date=exp)[["Expiry","Option Type","Strike Price","Open","High","Low","Close"]]
                df2 = get_history(symbol="NIFTY",
                        start=x,index = True,
                        end=exp,
                        option_type="CE",
                        strike_price = strk,
                        expiry_date=exp)[["Option Type","Open","High","Low","Close","Underlying"]]
    
                #x = nifty_master['PDate'].shift(-1)[nifty_master.index == exp].astype(date).iloc[0]
                df3 = mergefunc(df1,df2)
                df3["Cost"] = df3["Open_x"].iloc[0] + df3["Open_y"].iloc[0]
                df3["Value"] = df3["Low_x"] + df3["Low_y"]
                df3["Max_Profit"] = (df3["Value"].max(axis=0)) - df3['Cost'] 
                df3["Max_DD"] = (df3["Value"].min(axis=0)) - df3['Cost'] 
                df3["Max_Profit_Pct"] = df3["Max_Profit"]/df3["Cost"]
                df3["Max_DD_Pct"] = df3["Max_DD"]/df3["Cost"]
                df = df.append(df3) 
            except:
                #x = nifty_master['PDate'].shift(-1)[nifty_master.index == exp].astype(date).iloc[0]
                print("Error downloading month ", i)  
                
df                
```
